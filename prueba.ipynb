```python
import requests
import json
import pandas as pd
from datetime import datetime

# URL of the API
url = "https://www.imhpa.gob.pa/es/niveles-data?estacion=1080105"

try:
    # Fetch data from the API
    response = requests.get(url)
    response.raise_for_status()  # Raise an exception for HTTP errors
    data = response.json()["data"]  # Parse JSON response

except requests.exceptions.RequestException as e:
    print(f"Error fetching data from API: {e}")
except json.JSONDecodeError as e:
    print(f"Error decoding JSON response: {e}")
except Exception as e:
    print(f"An error occurred: {e}")

data
```

```python
registros_estructurados = []
for registro in data :
    print(registro)
    registros_estructurados.append({
        "timestamp": registro[0],
        "dia": datetime.utcfromtimestamp(registro[0]),
        "nivel": registro[1]
    })

registros_estructurados
```
