```python
#clusterizacion K-Means
import pandas as pd
datos = pd.read_csv('./housing.csv')

X = datos.loc[:, ['latitude', 'longitude', 'median_income']]
X
```

```python
import seaborn as sb

sb.scatterplot(data=X, x='longitude', y="latitude", hue="median_income", palette='coolwarm')
```

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters=6)

predicciones = model.fit_predict(X)

X['cluster'] = predicciones
X
```

```python
sb.scatterplot(data=X, x='longitude', y="latitude", hue="cluster", palette='coolwarm')
```

```python
X.groupby("cluster").mean()
```

```python
X["cluster"].hist(bins=50, figsize=(20,15), edgecolor='black')
```
